```python
cod_munic = microdadosEnade['CO_MUNIC_CURSO'] #Pegue a coluna 'CO_MUNIC_CURSO' e jogue em 'cod_munic'
cod_munic.value_counts() # Conte a quantidade de valores únicos de 'cod_munic', quantas vezes aparece cada município.
```

```python
cod_munic.hist()# Histograma da quantidade de candidatos por cidade de curso
```

```python
nu_idade = microdadosEnade['NU_IDADE'] #Atribuímos a coluna 'NU_IDADE' na variável 'nu_idade'
nu_idade.value_counts().sort_index() 
# Pedimos a quantidade em que cada valor de idade aparece, e ordenamos crescente de idade. Verifique que há alguem de 4 anos e 2 pessoas de 94 anos
```

```python
nu_idade.hist(bins=74) 
#Pedimos um histograma nos nossos dados de idade, especifiquei o bins em 74 pois eu queria uma barrinha por idade
#Chama a atenção esses picos repetidos em ciclos de 5 anos aproximadamente.
```

```python
tp_sexo = microdadosEnade['TP_SEXO'] #Atribuímos a coluna 'TP_SEXO' na variável 'tp_sexo'
tp_sexo.value_counts(), tp_sexo.hist()# Pedimos a quantidade em que cada valor 'F' ou 'M' aparece, um histograma.
```

```python
co_turno_graduacao = microdadosEnade['CO_TURNO_GRADUACAO']#Atribuímos a coluna 'CO_TURNO_GRADUACAO' na variável 'co_turno_graduacao'
co_turno_graduacao.value_counts(), co_turno_graduacao.hist()
```

```python
temp_espera = microdadosEnade['ANO_IN_GRAD']-microdadosEnade['ANO_FIM_EM'] 
#Subtraímos o ano de início do curso pelo ano de início da graduação e jogamos esses valores em 'temp_espera'
temp_espera.value_counts() # Contamos os valores
#Verifique que devido a erros em nossa base surgiram valores absurdos como -195 ou acima de 100.
```

```python
modais = microdadosEnade['CO_MODALIDADE'].value_counts() 
#Solicitamos as quantidade de alunos por modalidade de ensino, 1 = Presencial e 2 = A distância, atribiíms a modal
modais #Solicitamos o modal
```

```python
percent_modais = [100 * modais/modais.sum()] 
#Multiplicamos cada valor de modal por 100 e dividimos pela soma dos valores pra obter o percentual, atribuímos a 'percent_modais'
percent_modais #Solicitamos percent_modais
```
